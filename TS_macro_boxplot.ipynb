```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("ts_telemetry.csv.gz")
```

```python
sample_df = df.query("total_scored in [5000,10000,50000,75000,100000]").copy()
```

```python
rg_list = ["R1","R2"]
sample_df['Substituent'] = [rg_list[i] for i in sample_df.cycle]
```

```python
ax = sns.boxplot(x="total_scored",y="location",hue='Substituent',data=sample_df,hue_order=["R1","R2"])
ax.set_ylabel("Average ROCS Score for Each Building Block")
ax.set_xlabel("Iterations")
ax.legend(loc="upper left", bbox_to_anchor=(1.00, 0.75), ncol=1);
```
